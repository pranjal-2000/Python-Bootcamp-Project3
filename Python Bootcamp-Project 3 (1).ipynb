```python
# Tictactoe game
import random
```

```python
def game():
    board=['adjustment to adjust the position from 1',' ',' ',' ',' ',' ',' ',' ',' ',' ']
    symbol=[('X','O'),('O','X')]
    input("Enter your marker(X/O):").upper()
    sym1,sym2=symbol[random.randint(0,1)]
    print(f'{sym1} will go first')
    create_board(board)
    user_input(board,sym1,sym2)
def create_board(board):
    print(f"""
             {board[7]}|{board[8]}|{board[9]}                                    7|8|9
            --|-|--                                  --|-|--
             {board[4]}|{board[5]}|{board[6]}       Positions---->               4|5|6 
            --|-|--                                  --|-|--
             {board[1]}|{board[2]}|{board[3]}                                    1|2|3
          """)
def user_input(board,sym1,sym2):
    for i in range(0,9):
        while True:
            position=input("Enter the position(1-9):")
            if position.isnumeric():
                pos=int(position)
                if board[pos]!=' ':
                    print("Sorry!This position is already filled and cannot be overwritten.Please enter a another position between 0-9")
                    continue
                else:
                    break
            if position.isnumeric()==False:
                print("Sorry!Please enter a valid position between 0-9")
            else:
                pos=int(position)
                break     
        while i%2==0:
            print(f'{sym1} chance')
            board[pos]=sym1
            create_board(board)
            turn=sym1
            break
        while i%2!=0:
            print(f'{sym2} chance')
            board[pos]=sym2
            create_board(board)
            turn=sym2
            break
        if i>=4:
            if check(board):
                create_board(board)
                print(f'Congrats!{turn} won the game.')
                break
            else:
                if i==8:
                    create_board(board)
                    print("Nobody won the game.It's a tie!")
                    break
def check(board):
    cond=True
    if board[1]==board[2]==board[3]!=' ' or board[4]==board[5]==board[6]!=' ' or board[7]==board[8]==board[9]!=' ':
        return cond
    elif board[1]==board[4]==board[7]!=' ' or board[2]==board[5]==board[8]!=' ' or board[3]==board[6]==board[9]!=' ':
        return cond
    elif board[1]==board[5]==board[9]!=' ' or board[3]==board[5]==board[7]!=' ':
        return cond
def main():
    response='Y'
    while response=='Y':
        game()
        response=input("Do you wish to continue playing? ('Y'/'N'):").upper()
   
        
main()        
        
```
